```python
import os
import zipfile
import pandas as pd

folder_path = r"C:\Users\ìœ¤ì„œí˜„\Desktop\ì£¼í™©ìœ„ì›ë‹˜ì¶”ê°€ì™¸ë¶€ì‚¬ë¡€ì™¸ë¶€ë°ì´í„°"
output_root = r"C:\Users\ìœ¤ì„œí˜„\Desktop\ê°€ë³´ì"

os.makedirs(output_root, exist_ok=True)

for file in os.listdir(folder_path):
    if file.lower().endswith(".zip"):
        zip_path = os.path.join(folder_path, file)
        zip_name = os.path.splitext(file)[0]

        output_folder = os.path.join(output_root, zip_name)
        os.makedirs(output_folder, exist_ok=True)

        with zipfile.ZipFile(zip_path, 'r') as z:
            file_list = z.namelist()

            train_files = [
                f for f in file_list
                if f.lower().startswith("train") and (f.lower().endswith(".csv") or f.lower().endswith(".xlsx"))
            ]

            if not train_files:
                print(f"âŒ {file} ì— train íŒŒì¼ ì—†ìŒ")
                continue

            train_file = train_files[0]
            ext = os.path.splitext(train_file)[1]
            save_path = os.path.join(output_folder, f"data{ext}")

            # ZIPì—ì„œ ì„ì‹œë¡œ ê·¸ëŒ€ë¡œ ì¶”ì¶œ
            with z.open(train_file) as source, open(save_path, "wb") as target:
                target.write(source.read())

            # --- ì—¬ê¸°ì„œ CSV ì¸ì½”ë”© ë¬¸ì œ í•´ê²° ---
            if ext == ".csv":
                try:
                    df = pd.read_csv(save_path, encoding='utf-8')  # í˜¹ì‹œ UTF-8ì´ë©´ ê·¸ëŒ€ë¡œ ì½í˜
                except:
                    df = pd.read_csv(save_path, encoding='cp949')   # ëŒ€ë¶€ë¶„ í•œêµ­ CSVëŠ” cp949

                # UTF-8ë¡œ ì¬ì €ì¥
                df.to_csv(save_path, index=False, encoding='utf-8-sig')

            print(f"âœ” {file} â†’ data{ext} ì €ì¥ ì™„ë£Œ: {save_path}")
```

```python
import os

root_path = r"C:\Users\ìœ¤ì„œí˜„\Desktop\ë¬¸ì œí•´ê²°ì€í–‰\ì™¸ë¶€ì‚¬ë¡€ì‘ì—…ì‚°ì¶œë¬¼ëª¨ìŒ"

def is_utf8(file_path):
    try:
        with open(file_path, "r", encoding="utf-8") as f:
            f.read()
        return True
    except:
        return False

sample_key = "data_sample"
column_key = "data_column"
output_keys = ["output1", "output2", "output3", "process"]

passed = 0
failed = 0

for current_path, dirs, files in os.walk(root_path):

    sample_files = [f for f in files if sample_key in f.lower()]
    column_files = [f for f in files if column_key in f.lower()]
    output_png_files = [
        f for f in files 
        if any(k in f.lower() for k in output_keys) and f.lower().endswith(".png")
    ]

    # í•´ë‹¹ í´ë”ê°€ ê²€ì‚¬ ëŒ€ìƒì´ ë  ë•Œë§Œ ì²˜ë¦¬
    if not (sample_files or column_files or output_png_files):
        continue

    print(f"\nğŸ“ í´ë”: {current_path}")

    # --- ì¡°ê±´ ì²´í¬ ---
    sample_ok = bool(sample_files)
    column_ok = bool(column_files)
    output_ok = bool(output_png_files)

    print(" - sample ì¡´ì¬:", sample_files if sample_ok else "ì—†ìŒ")
    print(" - column ì¡´ì¬:", column_files if column_ok else "ì—†ìŒ")
    print(" - output/process PNG:", output_png_files if output_ok else "ì—†ìŒ")

    # CSV UTF-8 ì—¬ë¶€ ì²´í¬
    encoding_ok = True
    for f in sample_files + column_files:
        file_path = os.path.join(current_path, f)

        if f.lower().endswith(".csv"):
            if is_utf8(file_path):
                print(f"   âœ” {f}: UTF-8 CSV")
            else:
                print(f"   âŒ {f}: UTF-8 ì•„ë‹˜")
                encoding_ok = False
        else:
            print(f"   â„¹ {f}: CSV ì•„ë‹˜ (xlsx ë“±)")

    # --- í†µê³¼ ì¡°ê±´ ---
    # 1. sample íŒŒì¼ ìˆìŒ
    # 2. column íŒŒì¼ ìˆìŒ
    # 3. output PNG íŒŒì¼ ìˆìŒ
    # 4. sample/column CSVëŠ” UTF-8ì´ì–´ì•¼ í•¨
    if sample_ok and column_ok and output_ok and encoding_ok:
        print(" ğŸ‘‰ ê²°ê³¼: âœ” PASS")
        passed += 1
    else:
        print(" ğŸ‘‰ ê²°ê³¼: âŒ FAIL")
        failed += 1

# ---- ìµœì¢… ê²°ê³¼ ----
print("\n======================")
print(f"ìµœì¢… PASS ê°œìˆ˜: {passed}")
print(f"ìµœì¢… FAIL ê°œìˆ˜: {failed}")
print("======================")
```
