### 신입생 충원율 및 경쟁률

```python
#엑셀 파일 불러오는 함수
import pandas as pd

def load_and_clean_data(filepath, header):
    df = pd.read_excel(filepath, header = header)

    new_columns = []
    for col in df.columns:
        merged_name = [str(c) for c in col if 'Unnamed' not in str(c)]
        new_columns.append(' '.join(merged_name).strip())

    df.columns = new_columns
    df= df.fillna(method='ffill', axis=0)

    return df
```

```python
#야간 제외
df_freshman_noon = df_freshman[~df_freshman['구분'].str.contains('야간', na=False)]
```

### 전임교원 강의 비율

```python
#파일 불러오기
df_lecture = load_and_clean_data(r"C:\Users\윤서현\Desktop\국립강릉원주대 AI STUDIO\데이터정리\학과경쟁력분석\전임교원 강의 비율\교원 강의 담당 비율 (대학)_2025-07-03134522784.xlsx", header=[3,4,5])
```

```python
#야간 제외하기
df_lecture[~df_lecture['구분'].str.contains('야간', na=False)]

#기준연도 정수로 변경
df_lecture['기준연도'] = df_lecture['기준연도'].astype(int)
```

```python
df_lecture.columns
```

```python
#필요한 열 선택
df_lecture_selected = df_lecture[['기준연도', '학교', '단과대학', '학과(전공)','학기', '전임교원 비율']].copy()
```

```python
def create_mean_column(df, semester, col_name):
    df_mean = df[df['학기'].isin(semester)]
    
    # 학교와 학과 단위로 그룹핑
    df_avg = df_mean.groupby(['학교', '단과대학','학과(전공)'])['전임교원 비율'].mean().reset_index().round(2)
    new_column_name = f'{col_name} 전임교원 비율 평균'
    df_avg = df_avg.rename(columns={'전임교원 비율': new_column_name})
    
    # 병합 기준도 동일하게
    df_result = pd.merge(df, df_avg, on=['학교', '단과대학', '학과(전공)'], how='left')
    return df_result

df_lecture_selected_calculated = create_mean_column(df_lecture_selected, ['1학기', '2학기'], '정규학기')
df_lecture_selected_calculated = create_mean_column(df_lecture_selected_calculated, ['여름계절학기', '겨울계절학기'], '계절학기')
df_lecture_selected_calculated = create_mean_column(df_lecture_selected_calculated, ['1학기', '2학기','여름계절학기', '겨울계절학기'], '전체학기')
```

```python
df_lecture_selected_calculated
```

```python
# 전임교원 데이터에서 필요한 열만 선택 후 그룹별 대표값 하나만 남기기
df_lecture_clean = df_lecture_selected_calculated.groupby(['학교', '단과대학','학과(전공)'], as_index=False)[
    ['정규학기 전임교원 비율 평균', '계절학기 전임교원 비율 평균', '전체학기 전임교원 비율 평균']
].first() 
```

```python
df_lecture_clean.head()
```

```python
df_lecture_clean.rename(columns={'학과(전공)': '학과'}, inplace=True)
```

```python
# '이수\n학생수(명) 타학과 계열 인문·사회', '이수\n학생수(명) 타학과 계열 자연과학','이수\n학생수(명) 타학과 계열 공학', '이수\n학생수(명) 타학과 계열 의학','이수\n학생수(명) 타학과 계열 예·체능'
for df in df_list:
    df['타계열학생'] = (df['이수\n학생수(명) 타학과 계열 인문·사회'] + df['이수\n학생수(명) 타학과 계열 자연과학'] + df['이수\n학생수(명) 타학과 계열 공학'] 
                    + df['이수\n학생수(명) 타학과 계열 의학'] + df['이수\n학생수(명) 타학과 계열 예·체능'])
```

```python
# '기준연도', '학교', '단과대학', '학과\n(전공)', '계열','구분','교과목명','학기','이수\n학생수(명) 해당학과','타계열학생'

def create_summarized_dataframe(df):
    # 필요한 컬럼만 선택
    summary_df = df[['기준연도', '학교', '단과대학', '학과\n(전공)', '계열','구분','교과목명','학기','이수\n학생수(명) 해당학과','타계열학생']].copy()
    return summary_df


summarized_list=[]
for df in df_list:
    summary = create_summarized_dataframe(df)
    summarized_list.append(summary)
```

```python
#주간만 살리기

df_summ_weekly=[]
for df in summarized_list:
    df_weekly = df[df['구분'] == '주간']
    df_summ_weekly.append(df_weekly)
```

```python
df_summ_weekly[19]
```

```python
result_list = []

for df in df_summ_weekly:
    # 대학 + 학과 기준으로 평균 계산
    df_avg = df.groupby(['학교', '단과대학','학과\n(전공)']).agg({
        '이수\n학생수(명) 해당학과': 'mean',
        '타계열학생': 'mean'
    }).reset_index()

    # 컬럼명 변경
    df_avg = df_avg.rename(columns={
        '이수\n학생수(명) 해당학과': '해당학과 평균 이수학생수',
        '타계열학생': '타계열 평균 이수학생수'
    })

    #반올림
    df_avg[['해당학과 평균 이수학생수', '타계열 평균 이수학생수']] = df_avg[[
    '해당학과 평균 이수학생수', '타계열 평균 이수학생수']].round(2)

    result_list.append(df_avg)

# 결과 모두 합치기
final_result = pd.concat(result_list, ignore_index=True)
```

```python
final_result['캡스톤디자인 평균 이수 학생 수'] = final_result['해당학과 평균 이수학생수'] + final_result['타계열 평균 이수학생수']
```

```python
final_result.drop(columns=['해당학과 평균 이수학생수', '타계열 평균 이수학생수'], inplace=True)
```

```python
# 특수기호 제거 + 공백 제거 함수 / 텍스트 정리
def clean_text_columns(df, columns):
    for col in columns:
        df[col] = df[col].astype(str)
        df[col] = df[col].str.replace('·', '', regex=False)
        df[col] = df[col].str.replace('・', '', regex=False)
        df[col] = df[col].str.replace('?', '', regex=False)
        df[col] = df[col].str.replace(r'\s*_(제\d+캠퍼스|분교)', r'_\1', regex=True)
        df[col] = df[col].str.strip()
    return df
```

```python
# 전처리 적용
df_freshman_selected_and_clean = clean_text_columns(df_freshman_selected, ['학교', '학과', '기준연도','단과대학'])
df_dropout_selected_and_clean = clean_text_columns(df_dropout_selected, ['학교', '학과', '기준연도','단과대학'])
```
