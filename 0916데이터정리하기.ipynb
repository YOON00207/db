```python
import re

def lift_priority_rows(df: pd.DataFrame, cols=None, keep_score=False):
    """
    cols: 우선순위를 검사할 컬럼 리스트. None이면 전 컬럼 검사
    keep_score: True면 매칭 개수 점수 컬럼('_priority_score')를 남김
    """
    if cols is None:
        cols = df.columns.tolist()

    # 찾을 키워드 패턴
    pattern = re.compile(r'(거점국립대|강원권|본교)')

    # 각 행의 우선순위 점수(등장 횟수 합)
    # 문자열 아닌 값/NaN을 대비해 안전 처리
    scores = None
    for i, c in enumerate(cols):
        s = (
            df[c]
            .astype(str)
            .str.count(pattern)         # 해당 컬럼에서의 등장 횟수
            .fillna(0)
        )
        scores = s if scores is None else (scores + s)

    # 한 번이라도 매칭되면 1, 아니면 0
    prio_flag = (scores > 0).astype(int)

    # 우선순위 높은 행(1)이 위로 오도록 정렬 (안정 정렬)
    out = (
        df
        .assign(_priority_flag=prio_flag, _priority_score=scores)
        .sort_values(by=["_priority_flag", "_priority_score"], ascending=[False, False], kind="stable")
    )

    # 점수를 남길지 여부
    if keep_score:
        return out.drop(columns=["_priority_flag"]).rename(columns={"_priority_score": "우선순위_점수"})
    else:
        return out.drop(columns=["_priority_flag", "_priority_score"])

# ------------------ 사용 예시 ------------------
# cols를 지정하지 않으면 모든 컬럼에서 키워드를 찾습니다.
# 특정 컬럼(예: ['학교', '지역', '비고'])만 보고 싶으면 cols 인자로 넘기세요.
# 결과는 '거점국립대/강원권/본교'가 포함된 행이 위로 옵니다.
# df_sorted = lift_priority_rows(df)                             # 전 컬럼 검사
df_sorted = lift_priority_rows(df_final_2, cols=['학교'])       # 일부 컬럼만 검사
# df_sorted = lift_priority_rows(df, keep_score=True)            # 매칭 개수도 보고 싶을 때
```
