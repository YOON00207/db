```python
#학과코드 저장하기
import requests
import xml.etree.ElementTree as ET

def fetch_major_list(api_key):
    url = 'https://www.career.go.kr/cnet/openapi/getOpenApi'
    page = 1
    per_page = 100
    total_count = 0
    major_list = []  # 학과 리스트 저장 변수 추가

    while True:
        params = {
            'apiKey': api_key,
            'svcType': 'api',
            'svcCode': 'MAJOR',
            'contentType': 'xml',
            'gubun': 'univ_list',
            'thisPage': str(page),
            'perPage': str(per_page)
        }

        response = requests.get(url, params=params)
        response.raise_for_status()

        root = ET.fromstring(response.content)
        contents = root.findall('.//content')

        if not contents:
            print(f"모든 페이지를 불러왔습니다. 총 {total_count}건 수집 완료.")
            break

        for idx, content in enumerate(contents, 1):
            l_class = content.findtext('lClass')           # 계열
            m_class = content.findtext('mClass')           # 대표 학과명
            major_seq = content.findtext('majorSeq')       # 학과 코드
            facil_name = content.findtext('facilName')     # 관련 학과명 (복수)

            total_count += 1

            major_list.append({  # 저장
                '계열': l_class,
                '대표학과명': m_class,
                '학과코드': major_seq,
                '관련학과': facil_name
            })

        page += 1  # 다음 페이지로 넘김

    return major_list  # 반환 추가

if __name__ == '__main__':
    API_KEY = '1582dd032626cb967647f7717d1b67fb'
    majors = fetch_major_list(API_KEY)
    print(f" 수집 완료. 학과 수: {len(majors)}")
```

```python
import requests
import xml.etree.ElementTree as ET

# 1. 학과 리스트
major_name = [
    '건설환경전공', '건설환경공학과', '생명화학공학전공', '세라믹신소재공학전공', '신소재・생명화학공학부',
    '신소재금속공학전공', '전자・반도체공학부', '전자・반도체공학부(반도체공학전공)', '전자・반도체공학부(전자공학전공)',
    '스마트공간정보전공', '스마트인프라공학부', '디지털미디어커뮤니케이션학부', '경영학과', '경제학과', '관광경영학과',
    '국제통상학과', '도시계획・부동산학과', '무역학과', '법학과', '자치행정학과', '회계학과', '수산생명의학과',
    '식물생명과학과', '식품영양학과', '해양바이오식품학과', '해양바이오식품학과 해양바이오전공', '해양바이오식품학과 해양식품융합전공',
    '해양생명과학과', '해양생태환경학과', '환경조경학과', '음악학과', '조형예술・디자인학과', '체육학과', '패션디자인학과',
    '교직과정부', '국어국문학과', '사학과', '영어영문학과', '일본학과', '중어중문학과', '철학과', '대기환경과학과',
    '데이터사이언스학과', '생물학과', '수학물리학부', '수학물리학부(물리・에너지전공)', '수학물리학부(수학전공)',
    '화학신소재학과', '자유전공학부', '치위생학과', '치의예과', '치의학과'
]


# 2. 학과 리스트 수집 함수
def fetch_major_list(api_key):
    url = 'https://www.career.go.kr/cnet/openapi/getOpenApi'
    page = 1
    per_page = 100
    total_count = 0
    major_list = []

    while True:
        params = {
            'apiKey': api_key,
            'svcType': 'api',
            'svcCode': 'MAJOR',
            'contentType': 'xml',
            'gubun': 'univ_list',
            'thisPage': str(page),
            'perPage': str(per_page)
        }

        response = requests.get(url, params=params)
        response.raise_for_status()

        root = ET.fromstring(response.content)
        contents = root.findall('.//content')

        if not contents:
            print(f"모든 페이지를 불러왔습니다. 총 {total_count}건 수집 완료.")
            break

        for idx, content in enumerate(contents, 1):
            l_class = content.findtext('lClass')
            m_class = content.findtext('mClass')
            major_seq = content.findtext('majorSeq')
            facil_name = content.findtext('facilName')

            total_count += 1

            major_list.append({
                '계열': l_class,
                '대표학과명': m_class,
                '학과코드': major_seq,
                '관련학과': facil_name
            })

        page += 1

    return major_list


# 3. 학과 상세 정보 조회 함수
def fetch_major_detail(api_key, major_seq):
    url = 'https://www.career.go.kr/cnet/openapi/getOpenApi'
    params = {
        'apiKey': api_key,
        'svcType': 'api',
        'svcCode': 'MAJOR_VIEW',
        'contentType': 'xml',
        'gubun': 'univ_list',
        'majorSeq': str(major_seq)
    }

    response = requests.get(url, params=params)
    if response.status_code == 200:
        root = ET.fromstring(response.content)

        job = root.findtext('.//job')
        qualifications = root.findtext('.//qualifications')
        enter_fields = root.findall('.//enter_field/content')

        detail_info = {
            '관련 직업': job if job else '없음',
            '관련 자격': qualifications if qualifications else '없음',
            '졸업 후 진출 분야': []
        }

        if enter_fields:
            for field in enter_fields:
                gradeuate = field.findtext('gradeuate')
                description = field.findtext('description')
                detail_info['졸업 후 진출 분야'].append(f"{gradeuate} - {description}")

        return detail_info

    else:
        print(f'Error: {response.status_code}')
        return None


# 4. 학과 상세 정보 출력 함수
def fetch_and_display_detail(api_key, major_seq):
    detail = fetch_major_detail(api_key, major_seq)
    if detail:
        print(f"   관련 직업: {detail['관련 직업']}")
        print(f"   관련 자격: {detail['관련 자격']}")
        print(f"   졸업 후 진출 분야:")
        if detail['졸업 후 진출 분야']:
            for field in detail['졸업 후 진출 분야']:
                print(f"    - {field}")
        else:
            print(f"    없음")


# 5. 스마트 검색 (정확 일치 > 관련 학과 포함)
def search_major_smart(api_key, major_list, search_name):
    # 1. 대표학과명에서 정확히 일치하는 경우 찾기
    exact_matches = [major for major in major_list if search_name == major['대표학과명']]

    if exact_matches:
        major = exact_matches[0]  # 첫 번째 학과만 출력
        print(f"\n[정확 일치] {major['대표학과명']} (학과 코드: {major['학과코드']})")
        print(f"   계열: {major['계열']}")
        print(f"   관련 학과: {major['관련학과']}")
        fetch_and_display_detail(api_key, major['학과코드'])
        return

    # 2. 관련학과에 포함되는 경우 찾기
    related_matches = [major for major in major_list if search_name in major['관련학과']]

    if related_matches:
        major = related_matches[0]  # 첫 번째 학과만 출력
        print(f"\n[관련 학과 포함] {major['대표학과명']} (학과 코드: {major['학과코드']})")
        print(f"   계열: {major['계열']}")
        print(f"   관련 학과: {major['관련학과']}")
        fetch_and_display_detail(api_key, major['학과코드'])
        return

    # 3. 둘 다 없으면
    print(f"\n'{search_name}'(이)가 포함된 학과를 찾을 수 없습니다.")


# 6. 전체 자동 검색 실행
if __name__ == '__main__':
    API_KEY = '1582dd032626cb967647f7717d1b67fb'
    majors = fetch_major_list(API_KEY)

    for name in major_name:
        print(f"\n========== '{name}' 검색 시작 ==========")
        search_major_smart(API_KEY, majors, name)
```
