### 신입생 충원율 및 경쟁률과 중도탈락률 병합하기

```python
#학교명 기준

#'강원대학교', '강원대학교 _제2캠퍼스', '경북대학교', '경상국립대학교', '국립강릉원주대학교',
#'국립강릉원주대학교 _제2캠퍼스', '부산대학교', '전남대학교', '전남대학교 _제2캠퍼스', '전북대학교',
#'제주대학교', '충남대학교', '충북대학교', '가톨릭관동대학교', '경동대학교', '경동대학교 _제3캠퍼스',
#'경동대학교 _제4캠퍼스', '상지대학교', '연세대학교(미래) _분교', '한라대학교', '한림대학교'
```

```python
import pandas as pd

# 파일 불러오기
df_freshman = pd.read_excel(r"C:\Users\윤서현\Desktop\국립강릉원주대 AI STUDIO\데이터정리\학과경쟁력분석\신입생충원율,신입생경쟁률\0715 신입생충원율경쟁률_기준연도 2023.xlsx")
df_dropout = pd.read_excel(r"C:\Users\윤서현\Desktop\국립강릉원주대 AI STUDIO\데이터정리\학과경쟁력분석\중도탈락률\0715 중도탈락률 재시도.xlsx")

# 열 이름 통일하기
df_freshman.rename(columns={'학과\n(모집단위)': '학과'}, inplace=True)
df_dropout.rename(columns={'학과\n(전공)': '학과'}, inplace=True)
```

```python
df_freshman.head()
```

```python
# 특수기호 제거 + 공백 제거 함수 / 텍스트 정리
def clean_text_columns(df, columns):
    for col in columns:
        df[col] = df[col].astype(str)
        df[col] = df[col].str.replace('·', '', regex=False)
        df[col] = df[col].str.replace('・', '', regex=False)
        df[col] = df[col].str.replace('?', '', regex=False)
        df[col] = df[col].str.strip()
    return df

# 전처리 적용
df_freshman = clean_text_columns(df_freshman, ['학교', '학과', '기준연도','단과대학'])
df_dropout = clean_text_columns(df_dropout, ['학교', '학과', '기준연도','단과대학'])

# 병합
df_merged = pd.merge(
    df_freshman,
    df_dropout,
    on=['학교', '학과', '기준연도','단과대학'],
    how='outer'
)
```

```python
# df_merged = df_merged.drop(columns=[col for col in df_merged.columns if '단과대학' in col])
```

```python
df_merged = df_merged.drop(columns = '기준연도')
```

### 전임교원강의비율 병합하기

```python
#전임교원강의비율 파일 불러오기
df_class = pd.read_excel(r"C:\Users\윤서현\Desktop\국립강릉원주대 AI STUDIO\데이터정리\학과경쟁력분석\전임교원 강의 비율\0715 강의비율 재시도.xlsx")
```

```python
#학과로 컬럼명 통일
df_class.rename(columns = {'학과(전공)' : '학과'}, inplace = True)
```

```python
#같은 단과대학명 통일
df_class['단과대학'] = df_class['단과대학'].replace('트리니티자유대', '트리니티자유대학')
```

```python
#학교 리스트 확인
df_class['학교'].unique()
```

```python
#텍스트 정리
df_class_clean = clean_text_columns(df_class, ['학교','학과','단과대학'])
```

```python
#학교명 통일
campus_mapping2 = {
    '강원대학교': '강원대학교_본교',
    '강원대학교 _제2캠퍼스': '강원대학교_제2캠퍼스',
    '경북대학교': '경북대학교_본교',
    '경상국립대학교': '경상국립대학교_본교',
    '국립강릉원주대학교': '국립강릉원주대학교_본교',
    '국립강릉원주대학교 _제2캠퍼스': '국립강릉원주대학교_제2캠퍼스',
    '부산대학교': '부산대학교_본교',
    '전남대학교': '전남대학교_본교',
    '전남대학교 _제2캠퍼스': '전남대학교_제2캠퍼스',
    '전북대학교': '전북대학교_본교',
    '제주대학교': '제주대학교_본교',
    '충남대학교': '충남대학교_본교',
    '충북대학교': '충북대학교_본교',
    '가톨릭관동대학교': '가톨릭관동대학교_본교',
    '경동대학교': '경동대학교_본교',
    '경동대학교 _제3캠퍼스': '경동대학교_제3캠퍼스',
    '경동대학교 _제4캠퍼스': '경동대학교_제4캠퍼스',
    '상지대학교': '상지대학교_본교',
    '연세대학교(미래) _분교': '연세대학교(미래)_분교',
    '한라대학교': '한라대학교_본교',
    '한림대학교': '한림대학교_본교'

}
```

```python
# 기존 딕셔너리를 자동 반전
campus_mapping2_reversed = {v: k for k, v in campus_mapping2.items()}
```

```python
# 매핑 적용
df_class_clean['표준화된_학교명'] = df_class_clean['학교'].map(campus_mapping2_reversed)
```

```python
# 전임교원 강의비율 데이터에서 필요한 열만 선택 후 그룹별 대표값 하나만 남기기
df_class_clean_plus = df_class_clean.groupby(['표준화된_학교명', '학과','단과대학'], as_index=False)[
    ['정규학기 전임교원 비율 평균', '계절학기 전임교원 비율 평균', '전체학기 전임교원 비율 평균']
].first() 
```

```python
#학교로 컬럼명 통일
df_class_clean_plus.rename(columns = {'표준화된_학교명': '학교'}, inplace = True)
```

```python
df_class_clean_plus.head()
```

```python
#병합하기
df_merged_final_before_one = pd.merge(
    df_merged_freshman_dropout_gradu_cap_research_result_and_money,
    df_class_clean_plus,
    on=['학교', '학과','단과대학'],
    how='outer'
)
```
