```python
#엑셀 파일 읽어올 함수
import pandas as pd

def load_and_clean_data(filepath):
    df = pd.read_excel(filepath, header= [3,4,5])

    #컬럼명 정리
    new_columns=[]
    for col in df.columns:
        merged_name = [str(c) for c in col if 'Unnamed' not in str(c)]
        new_columns.append(' '.join(merged_name).strip())

    df.columns = new_columns
    df= df.fillna(method='ffill', axis=0)
    return df
```

```python
def create_summarized_dataframe(df_gw):
    # 필요한 컬럼만 선택
    summary_df = df_gw[['학교','단과대학','학과\n(모집단위)','학과상태']].copy()
    return summary_df

df_col = [create_summarized_dataframe(df) for df in df_list]
```

```python
#학과로 컬럼명 변경
for df in df_col:
    if '학과\n(모집단위)' in df.columns:
        df.rename(columns={'학과\n(모집단위)': '학과'}, inplace=True)
```

```python
#학과로 컬럼명 변경
for df in df_col:
    if '학과상태' in df.columns:
        df.rename(columns={'학과상태': '학과상태_2024'}, inplace=True)
```

```python
df_col
```

```python
#텍스트 정리하기
def clean_text_columns(df, columns):
    for col in columns:
        df[col] = df[col].astype(str)
        df[col] = df[col].str.replace('·', '', regex=False)
        df[col] = df[col].str.replace('・', '', regex=False)
        df[col] = df[col].str.replace('?', '', regex=False)
        df[col] = df[col].str.strip()
    return df

for i in range(len(df_col)):
    df_col[i] = clean_text_columns(df_col[i], ['학교', '학과','단과대학'])
```
