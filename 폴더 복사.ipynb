```python
import os
import shutil
import pandas as pd

excel_path = r"C:\Users\윤서현\Desktop\문제해결은행\외부사례_레시피_BASE((R)개요 및 성과).csv"
source_root = r"C:\Users\윤서현\Desktop\문제해결은행\외부사례작업산출물모음"
target_root = r"C:\Users\윤서현\Desktop\문제해결은행\외부사례레시피_BASE_기준_파일생성4"

df = pd.read_csv(excel_path)
os.makedirs(target_root, exist_ok=True)

# (완) 자동 인식
def find_matched_folder(root, data_id):
    for folder in os.listdir(root):
        if folder.startswith(data_id):
            return os.path.join(root, folder)
    return None

# 엑셀 분석방법에서 핵심 키워드 추출
def extract_keyword(text):
    keywords = ["군집", "예측", "분류", "이상치", "시계열", "텍스트",'기초통계']
    for k in keywords:
        if k in text:
            return k
    return None

# 🔥 폴더 내부의 내용만 복사하는 함수
def copy_folder_contents(src, dst):
    os.makedirs(dst, exist_ok=True)

    for item in os.listdir(src):
        s = os.path.join(src, item)
        d = os.path.join(dst, item)

        if os.path.isdir(s):
            shutil.copytree(s, d, dirs_exist_ok=True)
        else:
            shutil.copy2(s, d)


for idx, row in df.iterrows():

    file_name  = str(row["파일명"]).strip()
    data_id    = str(row["데이터 식별번호"]).strip()
    raw_method = str(row["WF분석방법"]).strip()

    keyword = extract_keyword(raw_method)

    src_folder = find_matched_folder(source_root, data_id)
    if not src_folder:
        print(f"⚠ 폴더 없음: {data_id}")
        continue

    dst_folder = os.path.join(target_root, file_name)

    # 원본 폴더에서 "하위폴더만" 찾기 (파일 제외)
    subfolders = [
        f for f in os.listdir(src_folder)
        if os.path.isdir(os.path.join(src_folder, f))
    ]

    # ============================================
    # CASE 1: 하위폴더 없음 → 전체 복사
    # ============================================
    if len(subfolders) == 0:
        print(f"\n📌 {file_name}: 하위 폴더 없음 → 전체 복사")
        shutil.copytree(src_folder, dst_folder, dirs_exist_ok=True)
        continue

    # ============================================
    # CASE 2: 하위폴더 있음 → keyword 매칭된 폴더만 복사
    # ============================================
    matched = []
    if keyword:
        for sub in subfolders:
            if keyword in sub:    # contain 매칭
                matched.append(sub)

    if matched:
        print(f"\n📌 {file_name}: '{keyword}' 포함된 폴더 내용만 복사됨 → {matched}")

        # 🔥 대상 결과 폴더 생성
        os.makedirs(dst_folder, exist_ok=True)

        for mf in matched:
            src = os.path.join(src_folder, mf)
            dst = dst_folder  # 🔥 폴더 이름 없이 “내용만 복사”
            copy_folder_contents(src, dst)

        continue

    # ============================================
    # CASE 3: 매칭 없지만 하위폴더 있음 → 아무것도 복사하지 않음
    # ============================================
    print(f"\n⚠ {file_name}: '{keyword}' 폴더 없음 → 복사하지 않음")

print("\n🎉 작업 완료!")
```

```python
import os

root_path = r"C:\Users\윤서현\Desktop\문제해결은행\외부사례레시피_BASE_기준_파일생성"

def is_utf8(file_path):
    try:
        with open(file_path, "r", encoding="utf-8") as f:
            f.read()
        return True
    except:
        return False

sample_key = "data_sample"
column_key = "data_column"
output_keys = ["output1", "output2", "output3", "process"]

passed = 0
failed = 0
fail_list = []   # 🔥 FAIL 상세 기록용

for current_path, dirs, files in os.walk(root_path):

    sample_files = [f for f in files if sample_key in f.lower()]
    column_files = [f for f in files if column_key in f.lower()]

    output_found = {key: any(key in f.lower() and f.lower().endswith(".png") for f in files)
                    for key in output_keys}

    # 검사 대상이 아니면 skip
    if not (sample_files or column_files or any(output_found.values())):
        continue

    # 결과 판단
    sample_ok = bool(sample_files)
    column_ok = bool(column_files)
    outputs_ok = all(output_found.values())  # 4개 모두 True

    encoding_ok = True
    for f in sample_files + column_files:
        file_path = os.path.join(current_path, f)
        if f.lower().endswith(".csv"):
            if not is_utf8(file_path):
                encoding_ok = False

    utf8_ok = encoding_ok

    # PASS / FAIL
    if sample_ok and column_ok and outputs_ok and utf8_ok:
        passed += 1
    else:
        failed += 1

        # 🔥 FAIL 상세 사유 저장
        fail_reason = []

        if not sample_ok:
            fail_reason.append("❌ data_sample 없음")
        if not column_ok:
            fail_reason.append("❌ data_column 없음")

        for key in output_keys:
            if not output_found[key]:
                fail_reason.append(f"❌ {key}.png 없음")

        if not utf8_ok:
            fail_reason.append("❌ sample/column UTF-8 아님")

        fail_list.append({
            "folder": current_path,
            "issues": fail_reason
        })

# ---- 최종 출력 ----
print("\n======================")
print(f"최종 PASS 개수: {passed}")
print(f"최종 FAIL 개수: {failed}")
print("======================")

print("\n🔥 FAIL 상세 리스트")
for item in fail_list:
    print(f"\n📂 폴더: {item['folder']}")
    for reason in item["issues"]:
        print(f"   - {reason}")
```

```python
import os

root_path = r"C:\Users\윤서현\Desktop\문제해결은행\외부사례레시피_BASE_기준_파일생성4"

target_keywords = ["data_sample", "data_column", "output1", "output2", "output3", "process"]

for folder in os.listdir(root_path):
    folder_path = os.path.join(root_path, folder)
    if not os.path.isdir(folder_path):
        continue

    print(f"\n📁 처리 중: {folder}")

    for file in os.listdir(folder_path):
        file_path = os.path.join(folder_path, file)
        if not os.path.isfile(file_path):
            continue

        name, ext = os.path.splitext(file)

        # 키워드 찾기
        matched_keyword = None
        for keyword in target_keywords:
            if keyword in name:
                matched_keyword = keyword
                break

        if not matched_keyword:
            continue

        # 새 파일명: 폴더명_keyword.ext
        new_name = f"{folder}_{matched_keyword}{ext}"
        new_path = os.path.join(folder_path, new_name)

        # ❗ 기존 파일이 있어도 덮어쓰기 위해 삭제
        if os.path.exists(new_path):
            os.remove(new_path)

        # 파일명 변경
        os.rename(file_path, new_path)
        print(f"   ➜ {file} → {new_name}")

print("\n🎉 모든 파일 이름 변경 완료!")
```

```python
import os
import pandas as pd

# 경로 설정
excel_path = r"C:\Users\윤서현\Desktop\문제해결은행\외부사례_레시피_BASE((R)개요 및 성과).csv"
base_root = r"C:\Users\윤서현\Desktop\문제해결은행\외부사례레시피_BASE_기준_파일생성"

df = pd.read_csv(excel_path)

# 타겟변수 컬럼 이름 후보(파일마다 다를 수 있음 대비)
target_columns = ["타겟 변수 여부","타겟 변수"]

print("\n🔍 타겟변수 O 없는 파일 목록\n")

problem_files = []

for _, row in df.iterrows():

    file_name = str(row["파일명"]).strip()
    wf_method = str(row["WF분석방법"]).strip()

    # 1) 예측모델 또는 분류모델만 검사
    if not ("예측모델" in wf_method or "분류모델" in wf_method):
        continue

    folder_path = os.path.join(base_root, file_name)

    if not os.path.isdir(folder_path):
        print(f"⚠ 폴더 없음: {file_name}")
        continue

    # 2) data_column 파일 찾기
    data_column_file = None
    for f in os.listdir(folder_path):
        if "data_column" in f.lower() and f.lower().endswith((".csv", ".xlsx")):
            data_column_file = os.path.join(folder_path, f)
            break

    if not data_column_file:
        print(f"❌ data_column 없음 → {file_name}")
        problem_files.append(file_name)
        continue

    # 3) 파일 읽기
    try:
        if data_column_file.endswith(".csv"):
            df_col = pd.read_csv(data_column_file)
        else:
            df_col = pd.read_excel(data_column_file)
    except Exception as e:
        print(f"❌ 파일 읽기 오류 → {file_name} / {e}")
        problem_files.append(file_name)
        continue

    # 4) 타겟변수 컬럼 찾기
    target_col = None
    for col in df_col.columns:
        if col in target_columns:
            target_col = col
            break

    if not target_col:
        print(f"❌ 타겟변수 컬럼 없음 → {file_name}")
        problem_files.append(file_name)
        continue

    # 5) 타겟변수에 'O' 있는지 확인
    values = df_col[target_col].astype(str).str.upper().str.strip()

    if not any(v == "O" for v in values):
        print(f"❌ 타겟변수에 'O' 없음 → {file_name}")
        problem_files.append(file_name)

# ---- 요약 ----
print("\n===========================")
print("🔎 'O' 없는 파일 총 개수:", len(problem_files))
print("===========================")

for f in problem_files:
    print(" -", f)
```
